```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time_x=np.linspace(0,40,1000)
```

```python
d_y1=2.5*time_x
```

```python
d_y2=3*(time_x-5)
```

```python
fig,ax=plt.subplots()
plt.title("first_example from jon Krohn")
# credits Dr Jon Krohn phd
plt.xlabel("time in minutes")
plt.ylabel("distnace in km")
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(time_x,d_y1,c="green")
ax.plot(time_x,d_y2,c="red")
plt.axvline(x=30,color='purple',linestyle='--')
plt.axhline(y=75,color='purple',linestyle='--')
```
